#### extract http information

```python
ETH = 0
IP = 1
TCP = 2
df_data = []
i = 0
http_lst = []
for pkt in http:
#     print(i)
#     i = i+1
    if len(pkt.layers) < 3:
        continue
    if pkt[2].layer_name != 'tcp':
        continue
    
    d = {
        'timestamp': pkt.sniff_time.timestamp(),
        'dest_ip': pkt[IP].dst,
        'dest_port': pkt[TCP].dstport,
        'highest layer': pkt.highest_layer,
        'length': pkt.length,
#         'response_code': pkt.http.respnse_code,
#         'server': pkt.http.server
    }
    df_data.append(d)
    http_lst.append(pkt)
```
