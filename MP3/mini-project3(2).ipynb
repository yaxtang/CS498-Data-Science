```python
import pandas as pd
import pyshark
import fileinput
from pandas.io.json import json_normalize

import matplotlib.pyplot as plt
```

```python
http = pyshark.FileCapture('http.pcap', keep_packets=True)
http.load_packets()
```

```python
len(http)
```

```python
dns = pyshark.FileCapture('dns.pcap', keep_packets=True)
dns.load_packets()
```

```python
len(dns)
```

# Task 0

```python
df_http = []
for pkt in http:
    if 'HTTP' not in str(pkt.layers):
        continue
    uri = pkt.http.get('request_uri')
    content_type = pkt.http.get('content-type')
    if uri or content_type:
        d = {
            'timestamp': pkt.sniff_time.timestamp(),
            'length': int(pkt.length),
            'highest_layer': pkt.highest_layer,
            'query_page': uri,
            'content_type': content_type,
            'src_ip': pkt.ip.src,
            'dest_ip': pkt.ip.dst,
            'dest_port': pkt.tcp.dstport
        }
        df_http.append(d)
df_http = pd.DataFrame(df_http)
df_http.fillna(value=float('nan'), inplace=True)
df_http.to_csv('df_http.csv')
df_http
```

```python
df_dns = []
for pkt in dns:
    if 'DNS' not in str(pkt.layers):
        continue
    d = {
        'timestamp': pkt.sniff_time.timestamp(),
        'length': int(pkt.length),
        'highest_layer': pkt.highest_layer,
        'response_type': int(pkt.dns.flags_response),
        'src_ip': pkt.ip.src,
        'dest_ip': pkt.ip.dst,
        'dest_port': pkt.udp.dstport
    }
    df_dns.append(d)
df_dns = pd.DataFrame(df_dns)
df_dns.to_csv('df_dns.csv')
df_dns
```

```python
df_dns[df_dns['response_type']==1]['src_ip'].unique()
```

#### Because the kernel module is downloaded from 162.212.156.148, we know that is the malicious server. And 75.75.75.75 is the legitimate server.

### Plot a histogram of the length of DNS queries for the two DNS servers.
### Clearly label the two servers using different colors/styles.
### Use log scale for the count on the Y axis

```python
import numpy as np
attack = df_dns[df_dns['src_ip'] == '162.212.156.148']
atk = np.sort(attack['length'].values)
legit = df_dns[df_dns['src_ip'] == '75.75.75.75']
lgt = np.sort(legit['length'].values)
```

```python
print(atk)
print(lgt)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
plt.figure(figsize=(20,10))
data = atk
plt.yscale('log')
plt.xlabel("length",fontsize = 20)
plt.ylabel("count",fontsize =20)
bins = np.linspace(50,1200,50)
plt.hist(data,bins,label ="attack",color = 'blue',alpha = 0.7);
plt.hist(lgt,bins,label ="Legitimate",color = 'orange',alpha = 0.9);
plt.legend(fontsize = 20)
plt.tick_params(axis='both', which='major', labelsize=20)
```
