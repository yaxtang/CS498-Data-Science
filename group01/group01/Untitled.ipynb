```python

ind = np.arange(0,1,1)
# print(len(ind))
fig, ax = plt.subplots(figsize= (15,10))
rects1 = ax.bar(ind, [0,1,2], 0.2, color='b')
rects2 = ax.bar(ind+0.4, [3,2,3], 0.2, color='r')
plt.xlabel("number of bits")
plt.ylabel("frequency")
plt.title("frequency of the bit error > 1 base on different type of nodes")
ax.set_xticks(np.asarray(ind))
ax.legend(['XE Bit Error','XK Bit Error', 'Service Bit Error'])

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%.4f' % round(float(height),4),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
# ,rects2,rects3)
plt.show()
```
